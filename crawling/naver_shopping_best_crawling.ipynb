```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url ='https://search.shopping.naver.com/best/category/click?categoryCategoryId=50000210&categoryChildCategoryId=&categoryDemo=A00&categoryMidCategoryId=50000210&categoryRootCategoryId=50000003&period=P1D'
```

```python
r = requests.get(url)
```

```python
html = r.text
html
```

```python
#BeautifulSoup으로 html소스를 python객체로 변환하기
#태그 정보를 이용할 수 있게 한다
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup
```

```python
goods_list = soup.select('div.imageProduct_text_area__3T1Mo > div.imageProduct_title__3TsP1') #리스트로 반환
#goods_list = soup.select('div.imageProduct_title__3TsP1') #가장 마지막 태그만 불러와도 같은 결과
#soup.find_all('div',{"class":"imageProduct_title__3TsP1"}) #다른 함수 같은 결과
goods_list #\32 6101847522 > 
```

```python
len(goods_list) #스크롤다운 안한 상태에서 15개의 정보가 노출
```

```python
#리스트 쪼개기
print("++네이버 쇼핑 생활가전 순위++")
num = 0
for i in goods_list:
    num += 1
    print(num, i.string)
```
