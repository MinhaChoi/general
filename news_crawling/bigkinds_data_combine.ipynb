```python
import pandas as pd
import numpy as np
import os
import natsort 
```

```python
path = 'C:\\Users\\mh\\jupyter code\\metaverse_news'
file_list = os.listdir(path)
file_list_py = [file for file in file_list if file.endswith('.xlsx')] ## 파일명 끝이 .csv인 경우
```

```python
file_list_py
```

```python
file_list_py = file_list_py[4:]
file_list_py
```

```python
after_order_list = natsort.natsorted(file_list_py)
after_order_list
```

```python
## csv 파일들을 DataFrame으로 불러와서 concat
df = pd.DataFrame()
for i in after_order_list:
    data = pd.read_excel(i, index_col = 0)
    df = pd.concat([df,data])
    
df = df.reset_index(drop = True)
```

```python
df
```

```python
#중복제거
df_new = df.drop_duplicates(subset=None, keep='first', ignore_index=True)
df_new
```

```python
df_new.to_csv('빅카인즈_메타버스.csv', encoding = 'utf-8-sig')
```
