```python
import pandas as pd
import numpy as np
import os
import natsort 
from tqdm import tqdm
```

```python
df = pd.read_csv('뉴스url수집2.csv', encoding = 'utf-8-sig', index_col=0)
df
```

```python
#time drop시키기
df.drop('time', axis=1, inplace = True)
df
```

```python
#time2 위아래 비교해보고 같은 값이면 해당 값으로 치환하기
#건너뛰어서 일 수도 있으니 -2 +2도 해보기
for i in tqdm(range(len(df['time2']))):
    #날짜가 입력이 안된 것 행에 대해서
    if "면" in df.time2.iloc[i]:
        #위아래를 비교해보고 
        #같으면
        if df.time2.iloc[i-1] == df.time2.iloc[i+1] :
            df.time2.iloc[i] = df.time2.iloc[i-2]
        #같지 않으면
        else :
            pass

#몇 개 남았는지 세보기 
count = 0
for i in tqdm(range(len(df['time2']))):
    #날짜가 입력이 안된 것 행에 대해서
    if "전" in df.time2.iloc[i]:
        count+= 1
count
```
