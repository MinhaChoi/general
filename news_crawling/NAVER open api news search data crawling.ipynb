# NAVER open api news search data crawling

네이버에서 제공하는 open api로 뉴스 검색 데이터 크롤링하기

<목적 : 네이버 뉴스 크롤링하기>

- 네이버에서 제공하는 open API client 발급받아서 사용할 시 -> 본문 수집 불가능 & 1000개 제한 (옵션 설정으로 갯수 제어 불가능)

- 네이버 뉴스  검색결과 -> 본문 수집 불가능 & 4000개 제한 (but, 옵션 설정으로 갯수 제어 가능)

- 공공데이터포털 -> 본문 수집가능 but, 메타버스 검색어로 12000개 한계, 실제 메타버스 검색어로 발생한 뉴스는 15만개 (22.02.28기준)

```python
import requests
import json
import time   
import pandas as pd
```

```python
df = pd.DataFrame() #df 미리 만들어두기
```

```python
url = 'https://openapi.naver.com/v1/search/news?'
clientId = "a" #발급받은 clientid 일일 25000건 호출 가능
clientSecret = "a" #발급받은 clientsecret
header = {
    "X-Naver-Client-Id":clientId,
    "X-Naver-Client-secret":clientSecret,
}

start = 1
while start <= 100 : #start는 1000이 한계
    time.sleep(0.1)
    params = {'query':"메타버스" , 'display':100, 'start':start, 'sort':'date'}
    r = requests.get(url, params = params, headers = header)
    time.sleep(0.1)
    j = r.json() 
    newslist = j['items'] 
    df = df.append(newslist)
    start += 100
    time.sleep(0.1)

#print(json.loads(r.text))
```
